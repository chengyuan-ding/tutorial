```python
df['date'] = pd.to_datetime(df['context_timestamp'], unit ='s').apply(lambda x : x.date())
```

```python
df['date'].unique()
```

```python
import datetime
```

```python
recent3days = df[(df['date'] >= datetime.date(2018, 9, 22)) & (df['date'] <= datetime.date(2018, 9, 24))].loc[:, ['item_id', 'is_trade']]
```

```python
## since trade is 1, non-trade is 0, the tradeOverClick = sum(is_trade)/count(is_trade), which is grouped by (item_id)
recent3daysConversion = recent3days.groupby('item_id').sum()
recent3daysConversion['click'] = recent3days.groupby('item_id').count()
recent3daysConversion['tradeOverClick'] = recent3daysConversion['is_trade'] / recent3daysConversion['click']
recent3daysConversion = recent3daysConversion.drop(columns=['is_trade', 'click'])
recent3daysConversion = recent3daysConversion.reset_index()
## leftjoin by key of "item_id"
recent3daysConversion.head()
```

```python
result = pd.merge(df, recent3daysConversion, on= 'item_id', how='left')
result.tail(10)
```
